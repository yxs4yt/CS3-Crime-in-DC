```python
import pandas as pd

#Importing pandas for data handling
files = ['Crime_Incidents_in_2024.csv', 'Crime_Incidents_in_2025.csv', 'Crime_Incidents_in_2026.csv']

#Reading files and forcing CCN to string so the merge doesn't break
dfs = [pd.read_csv(f, dtype={'CCN': str}) for f in files]

#Stacking all years into a single table
dc_crime_data = pd.concat(dfs, ignore_index=True)

#Converting date strings into datetime objects for time-series analysis
dc_crime_data['START_DATE'] = pd.to_datetime(dc_crime_data['START_DATE'])

#Extracting time components to study daily and monthly patterns
dc_crime_data['month'] = dc_crime_data['START_DATE'].dt.month
dc_crime_data['hour'] = dc_crime_data['START_DATE'].dt.hour

#Flagging records that fall within the summer season (June, July, August)
#This directly supports the seasonality hypothesis
dc_crime_data['is_summer'] = dc_crime_data['month'].isin([6, 7, 8]).map({True: 'Summer', False: 'Non-Summer'})


print(f"Total incidents merged: {len(dc_crime_data)}")
print(dc_crime_data[['START_DATE', 'OFFENSE', 'month', 'hour', 'is_summer']].head())


#dc_crime_data.to_csv('DC_Crime_Merged_2024_2026.csv', index=False)
```
